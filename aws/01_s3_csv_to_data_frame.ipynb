## Tutorial to convert a CSV file from S3 into a pandas DataFrame

```python
import boto3
import pandas as pd

def read_s3_csv(file_key, bucket_name, region_name):

    s3_client = boto3.client("s3", region_name)

    response = s3_client.get_object(Bucket=bucket_name, Key=file_key)

    status = response.get("ResponseMetadata", {}).get("HTTPStatusCode")

    if status == 200:
        print(f"Successful S3 get_object response. Status - {status}")
        train_df = pd.read_csv(response.get("Body"))        
        return train_df

    else:
        print(f"Unsuccessful S3 get_object response. Status - {status}")
```

## Replace bucket name, region name, and file key for your data

```python
bucket_name = "rml-project-data-us-east-1" ## bucket name
region_name = "us-east-1" ## region name
file_key = "mayankkthrgmail/large_nodeyiape/train.csv" # file key

df = read_s3_csv(file_key=file_key, bucket_name=bucket_name, region_name=region_name)
df.describe()
```
